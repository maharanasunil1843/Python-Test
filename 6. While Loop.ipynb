```python
a = 1
b = 10

while a <= b:
    print(a)
    a = a + 1
```

while check conditions in very first place.
if the condition is true then only it will enter into the loop and will keep on doing it unless the condition holds true.
Hence, it is called a conditional loop.

```python
while a <= b:
    print(a)
    a = a + 2
else:
    print("Print this else block")
```

```python
a
```

As we can see above, because the value of a has become 11 which is making the while condition false. Hence, it is not entering inside the loop.

```python
a,b = 1,10

while a <= b:
    print(a)
    a = a + 2
else:
    print("Print this else block")
```

```python
a,b = 1,10

while a <= b:
    print(a)
    if a == 5:
        break
    a = a +1
else:
    print("Print this else block")
```

Hence, since break is being applied, the else block is not used and the loop gets terminated

```python
a,b = 1,10
while a <= b:
    print(a)
    if a == 3:
        continue
    a = a +1
```

continue throws control directly to the while loop resulting into an infinite loop. It won't allow anything below it to be executed.

```python
a,b = 1,10
while a <= b:
    
    a = a +1
    if a == 3:
        continue
    print(a)
   
```
